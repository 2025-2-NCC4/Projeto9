```python
import pandas as pd


df_lojas = pd.read_excel("AnalisePicMoney.xlsx", skiprows=2, usecols="A:B")
df_lojas.columns = ["Loja", "valor_compra"]

df_lojas["valor_compra"] = df_lojas["valor_compra"].map(lambda x: f"R$ {x:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))

print("üìä Vendas por Loja")
print(df_lojas)



df_cupons = pd.read_excel("AnalisePicMoney.xlsx", skiprows=2, usecols="G:H")
df_cupons.columns = ["tipo_cupom", "valor_compra"]
df_cupons = df_cupons.dropna()

df_cupons["valor_compra"] = df_cupons["valor_compra"].map(lambda x: f"R$ {x:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))

print("\nüìä Comparativo Tipo de Cupom")
print(df_cupons)



df_tempo = pd.read_excel("AnalisePicMoney.xlsx", skiprows=2, usecols="M:N")
df_tempo.columns = ["data_captura", "valor_compra"]
df_tempo = df_tempo.dropna()

df_tempo["data_captura"] = pd.to_datetime(df_tempo["data_captura"], errors="coerce")
df_tempo["valor_compra"] = df_tempo["valor_compra"].map(lambda x: f"R$ {x:,.2f}".replace(",", "X").replace(".", ",").replace("X", "."))

print("\nüìä An√°lise Temporal")
print(df_tempo)
```

```python
import pandas as pd

all_sheets_dict = pd.read_excel("Analise PicMoney.xlsx", sheet_name=None)

df = all_sheets_dict["PicMoney-Massa_de_Teste_com_Loj"]


df["data_captura"] = pd.to_datetime(df["data_captura"], errors="coerce")


soma_por_dia = df.groupby("data_captura").agg(
    valor_total=("valor_compra", "sum"),
    quantidade_compras=("valor_compra", "count")
).reset_index()


soma_por_dia = soma_por_dia.sort_values("data_captura")


soma_por_dia["valor_total"] = soma_por_dia["valor_total"].map(lambda x: f"R$ {x:,.2f}")

print("üìà Soma dos valores de compra por dia + quantidade de compras:")
print(soma_por_dia)


```

```python
import pandas as pd

all_sheets_dict = pd.read_excel("Analise PicMoney.xlsx", sheet_name=None)

df = all_sheets_dict["PicMoney-Base_Simulada_-_Pe (2)"]


contagem = df.groupby(["local", "modelo_celular"]).size().reset_index(name="contagem")


mais_usados_por_local = contagem.loc[contagem.groupby("local")["contagem"].idxmax()].reset_index(drop=True)

print("üìç Modelo de celular mais usado por local:")
print(mais_usados_por_local)


contagem_tipo = df["tipo_celular"].value_counts().reset_index()
contagem_tipo.columns = ["tipo_celular", "contagem"]

percentuais_tipo = (contagem_tipo["contagem"] / contagem_tipo["contagem"].sum() * 100).round(2)
contagem_tipo["percentual"] = percentuais_tipo

print("\nüìà Percentual por tipo de celular:")
print(contagem_tipo)
```

```python
import pandas as pd

all_sheets_dict = pd.read_excel("Analise PicMoney.xlsx", sheet_name=None)

df = all_sheets_dict["PicMoney-Base_de_Transa√ß√µes_-_C"]


df["data"] = pd.to_datetime(df["data"], errors="coerce").dt.date


soma_por_dia = df.groupby(["data", "tipo_cupom"]).agg(
    valor_total=("valor_cupom", "sum"),
    quantidade_cupons=("valor_cupom", "count")
).reset_index()


tabela = soma_por_dia.pivot_table(
    index="data",
    columns="tipo_cupom",
    values="valor_total",
    aggfunc="sum",
    fill_value=0,
    margins=True,
    margins_name="Total Geral"
).reset_index()


for col in tabela.columns:
    if col not in ["data"]:
        tabela[col] = tabela[col].map(lambda x: f"R$ {x:,.2f}")

print("üìä Valores de cupom por dia e por tipo:")
print(tabela)
```
